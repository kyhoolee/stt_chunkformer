```python
import torch
from chunkformer_vpb.training.finetune_config import FinetuneConfig
from chunkformer_vpb.training.data_loader import get_dataloaders_debug

# Load config (dùng đường dẫn config YAML chuẩn của bạn)
CFG_PATH = "../../config/vpb_finetune_config.yaml"

cfg = FinetuneConfig.from_yaml(CFG_PATH)
train_loader, valid_loader = get_dataloaders_debug(cfg)

print(f"📦 Tổng số sample trong debug train: {len(train_loader.dataset)}")

# Lặp qua một batch đầu tiên để kiểm tra
for batch in train_loader:
    feats, feat_lens, toks, tok_lens, utt_ids, golds, preds = batch
    print(f"\n📌 Số lượng sample trong batch: {len(utt_ids)}")
    for i in range(len(utt_ids)):
        print(f"\n🧾 Sample {i+1}")
        print(f"🔑 utt_id          : {utt_ids[i]}")
        print(f"🗣️  text (json gốc): {cfg.tokenizer.tokenizer.decode_ids(toks[i][:tok_lens[i]].tolist())}")
        print(f"✅ gold_corrected  : {golds[i]}")
        print(f"🔁 pred_old        : {preds[i]}")
    break  # chỉ in 1 batch đầu tiên
```
