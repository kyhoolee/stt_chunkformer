```python
from datasets import load_dataset

ds = load_dataset("AILAB-VNUHCM/vivos", split="test", trust_remote_code=True)
```

```python
sample = ds[0]
print("WAV path:", sample["audio"]["path"])
print("Ground truth:", sample["sentence"])
print("Sample path:", sample["path"])
```

```python
for i in range(5):  # test thử 5 file đầu
    sample = ds[i]
    print(f"[{i}] WAV path:", sample["audio"]["path"])
    print(f"[{i}] Ground truth:", sample["sentence"])
    print()
```
